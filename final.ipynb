```python
import pandas as pd
from datetime import datetime
import re
import requests
from bs4 import BeautifulSoup
import hashlib as hl
import os
```

```python
data = pd.read_csv(r'/Users/xiaohengzhu/Desktop/2024 HDML/customer_information.csv')
data
```

```python
def calculate_age(birthdate):
    return datetime.now().year - int(birthdate.split('/')[2])

data['age'] = data['birthdate'].apply(calculate_age)
data.drop('birthdate', axis=1, inplace=True)
data.head()
```

```python
def delete_spaces(string):
    return string.replace(' ', '')

data['national_insurance_number'] = data['national_insurance_number'].apply(delete_spaces)
data.head()
```

```python
def categorise_age(age):
    if age < 18:
        return '[0, 18)'
    if age < 30:
        return '[18, 30)'
    if age < 50:
        return '[30, 50)'
    if age < 50:
        return '[40, 50)'
    if age < 60:
        return '[50, 60)'
    return '[60, 100)'

data['age_categories'] = data['age'].apply(categorise_age)
data.head()
```

```python
def get_area_from_postcode(postcode):
    
    def extract_letters_before_first_number(string):
        match = re.match(r'^\D+', string)
        if match:
            return match.group()
        else:
            return None
        
    outcode = postcode.split()[0]
    first_letters = extract_letters_before_first_number(outcode)
    return postcode_areas[first_letters]

url = "https://ideal-postcodes.co.uk/guides/postcode-areas"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
postcode_areas = {}

tables = soup.find_all('table')
for table in tables:
    for row in table.find_all('tr')[1:]:
        cells = row.find_all('td')
        if len(cells) > 1:
            area_code = cells[0].text.strip()
            area_name = cells[2].text.strip()
            postcode_areas[area_code] = area_name

data['city'] = data.apply(lambda row: get_area_from_postcode(row['postcode']), axis=1)
data.drop('postcode', axis=1, inplace=True)
data
```

```python
def group_area(area):
    if area in ['East Midlands', 'West Midlands']:
        return 'Midlands'
    if area in ['Scotland', 'Wales', 'Northern Ireland']:
        return 'Scotland, Wales & Northern Ireland'
    if area in ['North West', 'North East', 'Isle of Man']:
        return 'North of England'
    if area in ['South East', 'South West', 'Channel Islands', 'East of England', 'East England']:
        return 'South of England'
    if area == 'Greater London':
        return 'Greater London'

data['area_of_the_uk'] = data['city'].apply(group_area)
data.head()
```

```python
def categorise_education_level(education_level):
    if education_level == 'primary' or education_level == 'secondary':
        return 'primary or secondary'
    return 'bachelor or higher'

data['education_level_categories'] = data['education_level'].apply(categorise_education_level)
data.head()
```

```python
def calculate_BMI(weight, height):
    BMI = weight / (height ** 2)
    if BMI < 18.5:
        return 'Underweight'
    if BMI < 25:
        return 'Healthy weight'
    if BMI < 30:
        return 'Overweight'
    return 'Obese'

data['BMI'] = data.apply(lambda row: calculate_BMI(row['weight'], row['height']), axis=1)
data.head()
```

```python
def get_string_for_hash(first_name, surname, national_insurance_number):
        halved_ni = ""
        for index, char in enumerate(national_insurance_number):
            if index % 2 == 1:
                halved_ni += char
        return first_name + surname + halved_ni
data['string_for_hash'] = data.apply(lambda row: get_string_for_hash(row['given_name'], row['surname'], row['national_insurance_number']), axis=1)

def encode_data(data):
    encoded_data = data.encode('utf-8')
    return encoded_data
    
data['encoded_data'] = data.apply(lambda row: encode_data(row['string_for_hash']), axis = 1)

def generate_salt():
    salt = os.urandom(96)
    return salt
data['salt'] = data.apply(lambda row: generate_salt(), axis=1)

def combine_data_with_salt(data, salt):
    hash_data_with_salt = data + salt
    return hash_data_with_salt
data['data_and_salt_ready'] = data.apply(lambda row: combine_data_with_salt(row['encoded_data'], row['salt']), axis = 1)

hash_object = hl.sha256()

def hash_my_data(x):
    hash_object.update(x)
    return hash_object.hexdigest()

data['id'] = data.apply(lambda row: hash_my_data(row['data_and_salt_ready']), axis = 1)
data.head()
```

```python
def calculate_k_anonymity(data, columns):
    groups = data.groupby(columns).size()
    df = pd.DataFrame(groups, columns=['Count'])
    return groups.min(), df['Count'].value_counts().sort_index(ascending=True)

columns_for_k_anon = ['age_categories', 'gender', 'area_of_the_uk', 'education_level_categories']
for col in columns_for_k_anon:
    print(data[col].value_counts())

k_anon, groups = calculate_k_anonymity(data, columns_for_k_anon)
print(k_anon)
print(groups)
```
